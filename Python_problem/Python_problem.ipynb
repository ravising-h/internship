```python
#### authorization by using consumer key and access_token

auth = tweepy.OAuthHandler('','')

auth.set_access_token('','')
```

```python
### accessing twitter API.

api = tweepy.API(auth)
```

```python
#### VARIABLE: Twitter_user = target user  that is MIDASIIITD
#### Targeting user by using get_user funnction.

Twitter_user = 'MIDASIIITD'

try:
  user = api.get_user(Twitter_user)
except:
  print('Authorization falied.Check permisssion key or no such user')
```

```python
### INTENTION
### extract all tweets of user
### user.timeline() gives 200 tweets at a time.
### This function retrieves all the tweets of user by running the function many times.

def retrieving_tweets(user,first_tweet_id,no_of_status):
  
  
  
  '''
  PARAMETERS : user  : api.get_user(Twitter_user)   Twitter_user = midasiiitd 
               first_tweet_id : id of first tweet of user.
               no_of_status : no of tweets by user.
               
  While function runs multiple times untill all the tweets are retrieved.
   
  '''
    
  no_of_tweets_ret = 1
  result_set_midas = user.timeline(count = 1,truncated = False)
  no_of_tweets_not_retrieved = no_of_status
  last_id = first_tweet_id

   
  while (no_of_tweets_not_retrieved != 0):
        
      result_set_midas_older = user.timeline(max_id = last_id,count = no_of_tweets_not_retrieved,truncated = False) #Getting 200 tweets of user
      no_of_tweets_ret_ = len(result_set_midas_older) # number of tweets retrieved in first batch
      no_of_tweets_not_retrieved -= no_of_tweets_ret_  # number of tweets left.
      result_set_midas = result_set_midas + result_set_midas_older # adding all the tweets retrieved.
      last_id = result_set_midas[-1]._json['id'] # last tweet id: to get next batch batch of tweets 
     
    
  return result_set_midas
```

```python
first_tweet_id = user.timeline(count = 1,truncated = False)[0]._json['id']
result = retrieving_tweets(user,first_tweet_id,no_of_status)
```

```python
### appending the data with json data.
data = []
for i in range(no_of_status):
  data.append(result[i]._json)
```

```python
### saving it in jsonlines files
with open('result.jsonlines','w') as fp:
    json.dump(data, fp)
fp.close
```

```python
### Parsing in json Files
with open('result.jsonlines','r') as fp:
    try:
        data = json.load(fp)
    except:
        print('JSON is not loading')
    fp.close
```

```python
coloumn = ['Date','Time','text','retweet_count','favorite_count']
```

```python
### 2d matrix which will store all tweet information
### rows : no of tweets
### coloumn : ['Date','Time','Text','number of retweet','Likes','No of media']
tweet = [['' for i in range(len(coloumn)+1)] for j in range(no_of_status)]
```

```python
### formatting the data in the recieved from json files
for i in range(no_of_status):
    for j in range(len(coloumn)-1):
      if j == 0:
        temp = data[i]['created_at']
        
        tweet[i][0] = temp[0:11] + temp[-4:]
        tweet[i][1] = temp[11:20]
        
      else:
        tweet[i][j+1] = data[i][coloumn[j+1]]
    try:
        tweet[i][-1] = len([data[i]['entities']['media']])
    except:
        tweet[i][-1] = 'None'
```

```python
coloumn = ['Date','Time','Text','number of retweet','Likes','No of media']
d = pd.DataFrame(tweet,columns= coloumn )
```
